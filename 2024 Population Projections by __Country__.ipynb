2024 Population Projections by **Country**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# read data
df = pd.read_csv('2024Populations.csv')
df
```

```python

pd.set_option('display.max_columns', None)
df.head()
```

```python
df.info()
```

```python
# check for null values
print(df.isnull().sum())
# drop the null row in 3rd column
df = df.dropna(subset=['TwoLetterID'])
```

```python
crltn = df.drop(columns=['country','TwoLetterID'])
sns.heatmap(crltn.corr(), annot=True, cmap='coolwarm', lw=1)
plt.show()
```

```python
df
```

plot ten most populated countries

```python
# most populated nations

top10 = df.loc[df['pop2024'].nlargest(10).index]
top1980 = df.loc[df['pop1980'].nlargest(10).index]

sns.barplot(x=top10['country'], y=top10['pop2024'], hue=top10['unMember'])
plt.title('Top 10 Most Populated Countries 2024')

plt.xticks(rotation=90)
plt.show()
top10
```

india population growth over the years

```python
# india population growth over the years
ind_data = df[df['country']== 'India']
nind_data = ind_data.iloc[:,4:11]
print('nind_data')

# plot barchart
plt.figure(figsize=(10, 6))
sns.barplot(x=nind_data.columns, y=nind_data.iloc[0])
plt.title('India Population Growth Over the Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

plot population density

```python
sns.scatterplot(x=df['landAreaKm'], y=df['Density_2024'], hue=df['unMember'])
plt.show()
```
